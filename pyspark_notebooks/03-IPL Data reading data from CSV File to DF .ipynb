## Let us Upload IPL Cricket dataset CSV file

- Different way of loading CSV file to DataFram

```python
from pyspark.sql import SparkSession

#Load File
file_path = 'Files/ipldata/ipl_summary_raw.csv'

# Read the CSV file into a DataFrame
df = spark.read.csv(file_path, header=True, inferSchema=True)

# Show the first few rows of the DataFrame
df.show(5)
```

```python
# to view data foramat nicely and we can display this output in DataFrame
display(df)
```

## Understanding the Schema
You can explore the schema of the DataFrame to understand the structure of the data.

```python
df.printSchema()
```

## Basic Operation

```python
# Select specific columns
df.select("info_city", "info_teams_1", "info_teams_2").show()
```

```python
# Show distinct values of a column
df.select("info_city").distinct().show()
```

```python
# Show distinct values of a column
df.select("info_city").distinct().show()
```

```python
# Show distinct values of a column
df.select("info_teams_1").distinct().show()
```

```python
# Filter rows based on a condition
df.filter(df["info_city"] == "Chennai").show()
```
