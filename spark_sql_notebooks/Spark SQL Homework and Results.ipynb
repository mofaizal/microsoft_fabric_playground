## Homework Suggestions:
1. Write a query to find the total sales in each region.
2. Count how many orders were placed by each customer.
3. Find all orders with a discount of more than 20% but less than 30%.
4. List the top 3 orders by profit.
5. Replace the category 'Snacks' with 'Quick Bites' in the result set.

```python
df = spark.read.format("csv").option("header","true").load("Files/salesdata/Supermart Grocery Sales - Retail Analytics Dataset.csv")
# df now is a Spark DataFrame containing CSV data from "Files/salesdata/Supermart Grocery Sales - Retail Analytics Dataset.csv".
display(df)
```

```python
# Register the DataFrame as a temporary SQL table
df.createOrReplaceTempView("orders")
```

## Total Sales in Each Region

```python
%%sql
SELECT Region, SUM(Sales) AS Total_Sales 
FROM orders 
GROUP BY Region;
```

## Count of Orders Placed by Each Customer

```python
%%sql
SELECT Customer_Name, COUNT(*) AS Total_Orders 
FROM orders 
GROUP BY Customer_Name;
```

## Orders with Discount Between 20% and 30%

```python
%%sql
SELECT * 
FROM orders 
WHERE Discount > 0.20 AND Discount < 0.30;
```

## Top 3 Orders by Profit


```python
%%sql
SELECT * 
FROM orders 
ORDER BY Profit DESC 
LIMIT 3;
```

## Replace 'Snacks' with 'Quick Bites' in the Result Set

```python
%%sql
SELECT Order_ID, Customer_Name, 
       REPLACE(Category, 'Snacks', 'Quick Bites') AS Updated_Category, 
       Sub_Category, City, Sales 
FROM orders;
```
